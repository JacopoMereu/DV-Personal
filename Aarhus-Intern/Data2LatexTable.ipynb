# 1a versione Footnotes

````python
def json_to_columnar_latex_with_counts(
    json_data,
    caption="Caption",
    font_size="",  # e.g. "\\small", "\\footnotesize"
    footnotes=None  # e.g., {"*": ("footnote text", "symbol for superscript")}
):
    if footnotes is None:
        footnotes = {}

    seen_footnotes = set()

    # Split into column groups based on "new-column"
    columns = []
    current = []
    for entry in json_data:
        if "new-column" in entry:
            columns.append(current)
            current = []
        else:
            current.append(entry)
    columns.append(current)

    def replace_footnotes(text):
        for symbol, (note, superscript_key) in footnotes.items():
            if symbol in text:
                if symbol not in seen_footnotes:
                    seen_footnotes.add(symbol)
                    text = text.replace(symbol, fr"\footnote{{{note}}}")
                else:
                    text = text.replace(symbol, fr"\textsuperscript{{\textit{{{superscript_key}}}}}")
        return text

    def generate_tabular_block(block):
        lines = []
        lines.append(r"\begin{tabular}[t]{l r r}")
        lines.append(r"\textbf{Metric} & \textbf{\#} & \textbf{\%} \\")
        lines.append(r"\toprule")
        for section in block:
            title = replace_footnotes(section['title'])
            lines.append(fr"\textbf{{{title}}} & & \\")
            for item in section["data"]:
                name = replace_footnotes(item["name"])
                count = item["row_value"]
                perc = str(item["perc_value"]) + r"\%"
                lines.append(fr"{name} & {count} & {perc} \\")
            lines.append(r"\midrule")
        if lines[-1] == r"\midrule":
            lines[-1] = r"\bottomrule"
        else:
            lines.append(r"\bottomrule")
        lines.append(r"\end{tabular}")
        return "\n".join(lines)

    tabulars = [generate_tabular_block(col) for col in columns]

    latex = [r"\begin{table}[htp]", r"\centering"]
    if font_size:
        latex.append(font_size)

    for tab in tabulars:
        latex.append(r"\begin{minipage}{0.48\textwidth}")
        latex.append(tab)
        latex.append(r"\end{minipage}")
        latex.append(r"\hfill")

    latex.append(fr"\caption{{{caption}}}")
    latex.append(r"\end{table}")

    return "\n".join(latex)
````

````python
def json_to_columnar_latex_with_counts(
    json_data,
    caption="Caption",
    font_size="",  # e.g. "\\small", "\\footnotesize"
    footnotes=None  # e.g., {"*": ("footnote text", "symbol for superscript")}
):
    if footnotes is None:
        footnotes = {}

    seen_footnotes = set()

    # Split into column groups based on "new-column"
    columns = []
    current = []
    for entry in json_data:
        if "new-column" in entry:
            columns.append(current)
            current = []
        else:
            current.append(entry)
    columns.append(current)

    def replace_footnotes(text):
        for symbol, (note, superscript_key) in footnotes.items():
            if symbol in text:
                if symbol not in seen_footnotes:
                    seen_footnotes.add(symbol)
                    text = text.replace(symbol, fr"\footnote{{{note}}}")
                else:
                    text = text.replace(symbol, fr"\textsuperscript{{\textit{{{superscript_key}}}}}")
        return text

    def generate_tabular_block(block):
        lines = []
        # lines.append(r"\begin{tabular}[t]{l r r}")
        lines.append(r"\begin{tabular}[t]{l r l}")
        lines.append(r"\textbf{Metric} & \textbf{\#} & \textbf{\%} \\")
        lines.append(r"\toprule")
        for section in block:
            title = replace_footnotes(section['title'])
            lines.append(fr"\textbf{{{title}}} & & \\")
            for item in section["data"]:
                name = replace_footnotes(item["name"])
                count = item["row_value"]
                perc_float = float(item["perc_value"]) / 100  # convert to 0–1 float
                perc_bar = fr"\Chart{{{perc_float:.3f}}}"
                perc_display = f"{item['perc_value']}\\%"
                lines.append(fr"{name} & {count} & {perc_bar}~{perc_display} \\")
            lines.append(r"\midrule")
        if lines[-1] == r"\midrule":
            lines[-1] = r"\bottomrule"
        else:
            lines.append(r"\bottomrule")
        lines.append(r"\end{tabular}")
        return "\n".join(lines)

    tabulars = [generate_tabular_block(col) for col in columns]

    latex = [r"\begin{table}[htp]", r"\centering"]
    if font_size:
        latex.append(font_size)

    for tab in tabulars:
        latex.append(r"\begin{minipage}{0.48\textwidth}")
        latex.append(tab)
        latex.append(r"\end{minipage}")
        latex.append(r"\hfill")

    latex.append(fr"\caption{{{caption}}}")
    latex.append(r"\end{table}")

    return "\n".join(latex)
````

````python
# Example usage

# Foonotes of your table. If you do NOT have any, just put empty dict {}
# K: tuple(V1, V2), where
#   K => the marker inside your data. e.g. "*" is inside "Design tools regularly used*"
#   V1 is the message you want to write in the footnote. e.g. "(Select all that apply). This added up exceeds 100\%." Beware of escaping latex special chracters like %
#   V2 is the footnote marker you want to show. If you don't want it different, put a string equal to the original marker K.
m_footnotes = {
    "*": ("(Select all that apply). This added up exceeds 100\%.", "a"),
    "**": ("(Select top 3). This added up exceeds 100\%.", "b")
}
# Caption of your table.
# You can inject latex commands like textbf, but you have to escape it twice (\\textbf)
m_caption="\\textbf{Study \\textit{pre1}.} Participant demographics and background relevant to voice-based 2D GUI prototyping, providing context for evaluating a voice-driven conversational interface for GUI creation and editing."

# Fontsize inside the table.
m_font_size="\\small" # \tiny \footnotesize ...
latex_code = json_to_columnar_latex_with_counts(
    data,
    caption=m_caption,
    font_size=m_font_size,
    footnotes=m_footnotes
)

# Select console output (ctrl+A), copy (ctrl+C) and paste it in your latex file :)
print(latex_code)
````

function generateNValuesWithSumEqualToS(n, sum) {
  if (n <= 0 || sum < n) {
    throw new Error("Invalid input: sum must be at least n, and n must be > 0");
  }

  const values = [];
  let remainingSum = sum;

  for (let i = 0; i < n - 1; i++) {
    const maxVal = remainingSum - (n - i - 1); // ensure at least 1 remains per value
    const value = Math.floor(Math.random() * (maxVal - 1 + 1)) + 1;
    values.push(value);
    remainingSum -= value;
  }

  values.push(remainingSum);

  const normalized = values.map(v => v / sum);

  return {
    values,
    normalized
  };
}
