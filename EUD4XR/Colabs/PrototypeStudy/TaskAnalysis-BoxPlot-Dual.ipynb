```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

vr_score = np.array(
   [
    [6, 6, 6, 5, 6, 6],  # V1
    [6, 5, 1, 1, 5, 1], # A1 (None per dato mancante)
    [1, 5, 6, 5, 3, 6],  # A3
    [6, 6, 6, 4, 6, 6],  # A4
    [4, 6, 4, 1, 6, 6],  # A5
    [6, 6, 6, 5, 5, 6],  # A6
    [6, 6, 6, 6, 5, 6],  # A7
    [5, 6, 4, 6, 5, 6],  # A8
    [6, 6, 1, 6, 3, 3],  # J2
    [6, 4, 6, 6, 6, 6],  # A2
    [6, 4, 6, 6, 6, 2],  # A9
    [6, 6, 6, 6, 4, 6]   # A10
])

ar_score = np.array([
    [6, 6, 6, 5, 6, 6],  # V1
    [6, 5, 1, 1, 5, 1], # A1 (None per dato mancante)
    [1, 5, 6, 5, 3, 6],  # A3
    [6, 6, 6, 4, 6, 6],  # A4
    [4, 6, 4, 1, 6, 6],  # A5
    [6, 6, 6, 5, 5, 6],  # A6
    [6, 6, 6, 6, 5, 6],  # A7
    [5, 6, 4, 6, 5, 6],  # A8
    [6, 6, 1, 6, 3, 3],  # J2
    [6, 4, 6, 6, 6, 6],  # A2
    [6, 4, 6, 6, 6, 2],  # A9
    [6, 6, 6, 6, 4, 6]   # A10
])

def normalize_counts(data):
    counts = np.array([(data == i).sum(axis=0) for i in range(1, 7)], dtype=float)
    normalized = counts / counts.sum(axis=0)
    return normalized

labels = ["1  (U$_wB$_w | U$_pB$_w | U$_wB$_p)", "2 (U$_wB$_p | U$_pB$_p)", "3 (U$_wB$_c)", "4  (U$_cB$_p)", "5 (U$_pB$_c)", "6 (U$_cB$_c)"]

vr_counts = normalize_counts(vr_score)
ar_counts = normalize_counts(ar_score)
colors = ["#D7191C", "#e4765e", "#FDAE61", "#FFFFBF", "#A6D96A", "#1A9641"]


# Plot settings
fig, ax = plt.subplots(figsize=(14, 3))
x = np.arange(6)  # 6 groups (questions)
bar_width = 0.40  # Adjusted for spacing

# Initialize bottom for stacking
bottom_vr = np.zeros(6)

space_between_bars = 0.05

# Plot VR bars
for i in range(6):
    bars = ax.bar(x - bar_width / 2 - space_between_bars / 2, vr_counts[i], width=bar_width,
                  color=colors[i], bottom=bottom_vr)  # Assign specific color

    for bar, val in zip(bars, vr_counts[i]):
        if val > 0.05:  # Avoid cluttering with too small values
            s = f'{val:.0%}'.lstrip("0")  # Remove leading zero
            ax.text(bar.get_x() + bar.get_width() / 2,
                    bar.get_y() + bar.get_height() / 2, s,
                    ha='center', va='center', fontsize=10, color='black')

    # Add label only for the top-most bar
    if i == 5:  # i == 6 corresponds to the last (top-most) bar
        for bar, val in zip(bars, vr_counts[i]):
            ax.text(bar.get_x() + bar.get_width() ,
                    bar.get_y() + bar.get_height() + 0.12,
                    "VR - AR", ha='center', va='top', fontsize=10, color='black')

    bottom_vr += vr_counts[i]  # Update bottom for stacking



# Plot AR bars
bottom_ar = np.zeros(6)
for i in range(6):
    bars = ax.bar(x + bar_width / 2 + space_between_bars / 2, ar_counts[i], width=bar_width,
                  color=colors[i], bottom=bottom_ar)  # Assign specific color

    for bar, val in zip(bars, ar_counts[i]):
        if val > 0.05:  # Avoid cluttering with too small values
            s = f'{val:.0%}'.lstrip("0")  # Remove leading zero
            ax.text(bar.get_x() + bar.get_width() / 2,
                    bar.get_y() + bar.get_height() / 2, s,
                    ha='center', va='center', fontsize=10, color='black')

    bottom_ar += ar_counts[i]  # Update bottom for stacking


qLabels = ["Task 1",
           "Task 2",
           "Task 3",
           "Task 4",
           "Task 5",
           "Task 6"]

# Labels and title
# ax.set_xlabel("Questions")
ax.set_ylabel(None)
ax.set_title("Task Success", fontsize=16, fontweight='bold', pad=20)
ax.set_xticks(x)
ax.set_xticklabels([f"{qLabels[i]}" for i in range(6)])
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

# Improved legend in a separate white area
from matplotlib.patches import Patch
legend_patches = [Patch(color=colors[i], label=labels[i]) for i in range(6)]
plt.subplots_adjust(bottom=0.25)
legend_ax = fig.add_axes([0.1, 0.0, 0.8, 0.1], frameon=False)
legend_ax.set_xticks([])
legend_ax.set_yticks([])
legend_ax.legend(handles=legend_patches, loc='center', ncol=6)

plt.savefig('task_success.pdf', format='pdf', bbox_inches='tight')
# Show plot
plt.show()
```

```python
# prompt: I want you to show me a table with the mean and stdev of the frame df, grouping for values in the task column and  considering only the values where Scenario is VR. Convert the value that is in seconds into minutes and seconds

# Assuming df is already defined as in the provided code

# Filter for VR scenarios
vr_df = df[df["Scenario"] == "VR"]

# Group by task and calculate mean and standard deviation
summary_df = vr_df.groupby("Task")["Value"].agg(["mean", "std"]).reset_index()

# Convert seconds to minutes and seconds
def seconds_to_minutes_seconds(seconds):
    minutes = int(seconds // 60)
    seconds = int(seconds % 60)
    return f"{minutes}m {seconds}s"

summary_df["mean"] = summary_df["mean"].apply(seconds_to_minutes_seconds)
# Standard deviation is not directly converted to a time representation since it is a measure of spread.
# You can keep it as it is, or calculate the min/max of the range.
# For instance, you can use the mean value in seconds for calculating a range:

# Convert mean to seconds for range calculation
def minutes_seconds_to_seconds(time_str):
  parts = time_str.split("m ")
  minutes = int(parts[0])
  seconds = int(parts[1].rstrip('s'))
  return minutes * 60 + seconds


summary_df["mean_sec"] = summary_df["mean"].apply(minutes_seconds_to_seconds)
summary_df["min_range"] = summary_df["mean_sec"] - summary_df["std"]
summary_df["max_range"] = summary_df["mean_sec"] + summary_df["std"]


summary_df["min_range"] = summary_df["min_range"].apply(seconds_to_minutes_seconds)
summary_df["max_range"] = summary_df["max_range"].apply(seconds_to_minutes_seconds)


# Display the table
print(summary_df[["Task", "mean", "std", "min_range", "max_range"]])
```
